I noticed that most of the playing cards have different numbers of chips placed next to them, which might represent some kind of 'offset'. For example, the flushed face corresponds to the Four of Clubs with one chip next to it, and the Five of Hearts with no chips. Shifting 'S' and 'H' by 1 and 0 positions respectively gives 'T' and 'H'.
