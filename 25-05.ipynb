### Fibonacci

```python
def get(n):
    if n < 1:
        return []
    result = []
    seen = set()
    a, b = 1, 1  
    while True:
        num_digits = len(str(a))
        if num_digits == n:
            if a not in seen:
                result.append(a)
                seen.add(a)
        elif num_digits > n:
            break 
        a, b = b, a + b  
    return result

print(get(2))  
print(get(3))
print(get(4))
print(get(5))
print(get(6))
print(get(7))
print(get(8))
print(get(9))
print(get(10))
print(get(11))
```

### square

```python
import math

def get_squares(n):
    if n < 1:
        return []
    
    lower = 10 ** (n-1)
    upper = 10 ** n - 1
    
    start = math.ceil(math.sqrt(lower))
    end = math.floor(math.sqrt(upper))
    
    return [i*i for i in range(start, end + 1)]

print(get_squares(1)) 
print(get_squares(2))  
print(get_squares(3))  
print(get_squares(4))  
```

### 13 multiple

```python
import math

def get_multiples_of_13_by_digits(n):
    if n < 1:
        return []
    
    min_num = 10 ** (n-1)   
    max_num = 10 ** n - 1    
    
    if min_num > max_num:
        return []
    
    first = math.ceil(min_num / 13) * 13    
    last = math.floor(max_num / 13) * 13   
    return list(range(first, last + 1, 13)) if first <= max_num else []

print(get_multiples_of_13_by_digits(2)) 
print(get_multiples_of_13_by_digits(3)) 
print(get_multiples_of_13_by_digits(4))  
```

### prime

```python
import math

def generate_primes_by_digits(n):
    if n < 1:
        return []
    
    # 计算范围边界
    lower_bound = 10**(n-1)
    upper_bound = 10**n - 1
    
    # 处理小范围的边界情况
    if upper_bound < 2:
        return []
    
    # 计算筛法的上限（取平方根优化）
    sieve_limit = int(math.isqrt(upper_bound)) + 1
    
    # 初始化标记数组（只考虑奇数）
    is_prime = [True] * ((upper_bound + 1) // 2)
    
    # 处理1位数的特殊情况
    if n == 1:
        primes = [2, 3, 5, 7]
        return [p for p in primes if lower_bound <= p <= upper_bound]
    
    # 应用埃拉托斯特尼筛法（只处理奇数）
    for i in range(3, sieve_limit, 2):
        if is_prime[i // 2]:
            start = i*i
            step = 2*i  # 跳过偶数
            # 确保起始点是奇数
            if start % 2 == 0:
                start += i
            for j in range(start, upper_bound + 1, step):
                if j % 2 == 1:  # 只标记奇数
                    is_prime[j // 2] = False
    
    # 收集范围内的质数
    primes = []
    if n == 1 and lower_bound <= 2 <= upper_bound:
        primes.append(2)
    
    start_idx = max(3, lower_bound)
    if start_idx % 2 == 0:  # 确保从奇数开始
        start_idx += 1
        
    for num in range(start_idx, upper_bound + 1, 2):
        if num == 1:
            continue
        if num == 2 or is_prime[num // 2]:
            primes.append(num)
    
    return primes

# 示例用法
print(generate_primes_by_digits(1))  # 返回所有1位数质数: [2, 3, 5, 7]
print(generate_primes_by_digits(2))  # 返回所有2位数质数: [11, 13, ..., 97]
print(generate_primes_by_digits(3))  # 返回所有3位数质数: [101, 103, ..., 997]
```
